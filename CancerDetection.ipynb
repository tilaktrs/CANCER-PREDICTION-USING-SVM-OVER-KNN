```python
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
import pandas as pd#high level manipulation tool ,we can manipulate tabular data in rows of observations and columns of variables.
import seaborn as sns#Seaborn is a Python data visualization library based on matplotlib It provides a high-level interface for drawing attractive and informative statistical graphics
from sklearn.datasets import load_breast_cancer#inbuilt data
```

```python
cancer = load_breast_cancer()
```

```python
print(cancer.DESCR)#description of cancer dataset
```

```python
df_cancer = pd.DataFrame(np.c_[cancer['data'] , cancer['target']], columns=np.append(cancer['feature_names'],['target']))
df_cancer.head()
```

```python
sns.pairplot(df_cancer,hue='target',vars=['mean radius','mean texture','mean perimeter','mean area','mean smoothness','mean symmetry','mean fractal dimension','mean concavity'])
# orange = no cancer
# blue = cancer
#pair plotting means graph of one feature with respect to all other feature here all graph are drawn with respect to means radius
```

```python
X,y=load_breast_cancer(return_X_y=True)
```

```python
#train test split-Split arrays or matrices into random train and test subsets
```

```python
X_train, X_test, y_train, y_test = train_test_split(
...     X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
model.fit(X_train ,y_train)
```

```python
model.predict(X_test[:10])#predict first 10 people data whether cancer or not
```

```python
y_test[:10]
```

```python
model.score(X_test,y_test)
```
